```python
from sklearn.datasets import make_moons
import numpy as np
import matplotlib.pyplot as plt

x,y=make_moons(n_samples=500,
               noise=0.1,
               random_state=0)
plt.scatter(x[:,0],x[:,1],c=y)
y
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(100,50,10))
model.fit(x,y)

x1,x2=np.meshgrid(np.arange(-1,2,0.1),np.arange(-0.75,1.25,0.05))
x_grid=np.c_[x1.flatten(),x2.flatten()]
y_grid=model.predict(x_grid)

plt.scatter(x_grid[:,0],x_grid[:,1],c=y_grid,alpha=0.2)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
x=np.arange(-10,10,0.0001).reshape(-1,1)
y=x**2
plt.scatter(x,y)
```

```python
from sklearn.neural_network import MLPRegressor
model=MLPRegressor(hidden_layer_sizes=(200,10))
model.fit(x,y)

y_pred=model.predict(x)
plt.scatter(x,y)
plt.scatter(x,y_pred)
```
