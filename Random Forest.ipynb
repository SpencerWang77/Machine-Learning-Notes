```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
x = cancer['data']
y = cancer['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train,y_train
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
np.mean(model.predict(x_test)==y_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plot_tree(model,feature_names=cancer['feature_names'],
                  class_names=cancer['target_names'],
                  filled=True)
```

```python
c23=np.sort(np.unique(x_train[:,23]))
values=[]
for i in range(len(c23)-1):
  mid=(c23[i]+c23[i+1])/2
  group1=y_train[x_train[:,23]<mid]
  a=np.mean(group1)
  gini=1-a**2-(1-a)**2
  value=gini*len(group1)/len(y_train)
  group2=y_train[x_train[:,23]>mid]
  a=np.mean(group2)
  gini=1-a**2-(1-a)**2
  value2=gini*len(group2)/len(y_train)
  values.append((value+value2,mid))
  #print(mid)
values.sort()
values
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
np.mean(model.predict(x_test)==y_test)
```

```python
indices = np.random.choice(np.arange(100), 3, replace=False)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, idx in enumerate(indices):
  plot_tree(model.estimators_[i],
                  feature_names=cancer['feature_names'],
                  class_names=cancer['target_names'],
                  filled=True,
                  ax = axs[i])
```

```python
#random forest
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
model=DecisionTreeClassifier(max_features="sqrt")
model.fit(x_train,y_train)
np.mean(model.predict(x_test)==y_test)
```

```python
def randomforest(n):
  models=[]
  for i in range(n):
    indices=np.random.choice(np.arange(x_train.shape[0]),x_train.shape[0],replace=True)
    model=DecisionTreeClassifier(max_features="sqrt")
    model.fit(x_train[indices],y_train[indices])
    models.append(model)
  return models

models=randomforest(100)
results=np.zeros(x_test.shape[0])
for model in models:
  results+=model.predict(x_test)
results=results/x_test.shape[0]>=0.5

np.mean(y_test==results)

#x_test.shape,results.shape


```
