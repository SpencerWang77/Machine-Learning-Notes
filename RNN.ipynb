```python
import numpy as np
```

```python
steps=100
input_features=10
hidden_state=5
```

```python
inputs=np.random.normal(0,1,(steps,input_features))
state_T=np.zeros((hidden_state))
```

```python
w=np.random.normal(0,1,(hidden_state,hidden_state))
u=np.random.normal(0,1,(hidden_state,input_features))
b=np.random.normal(0,1,(hidden_state))
```

```python
res=[]
for i in range(steps):
  output=np.tanh(w@state_T+u@inputs[i]+b)
  res.append(output)
  state_T=output
res
```
